```python
import pandas as pd
import numpy as np

df = pd.read_csv('diamonds_cleaned.csv',encoding='utf-8')
```

```python
#Volume = x * y * z



df['volume'] = df['x'] * df['y'] * df['z']
```

```python
#Price per Carat = price / carat
df['price_per_carat_usd'] = df['price'] / df['carat']
```

```python
#Dimension Ratio = (x + y) / (2 * z)


df['dimension_ratio'] = (df['x'] + df['y']) / (2 * df['z'])
```

```python
#Carat Category: Light (<0.5), Medium (0.5-1.5), Heavy (>1.5)

def carat_category(c):

    if c < 0.5:
        return 'Light'
    elif c <= 1.5:
        return 'Medium'
    else:
        return 'Heavy'
df['carat_category'] = df['carat'].apply(carat_category)
```

```python
#Carat Category: Light (<0.5), Medium (0.5-1.5), Heavy (>1.5)

df[['carat','volume','price_per_carat_usd','carat_category']].head()
```

```python
#usd to inr conversion

usd_to_inr = 87.75   
df['price_inr'] = df['price'] * usd_to_inr
df['price_per_carat_inr'] = df['price_inr'] / df['carat']
print(f'Applied conversion: 1 USD = {usd_to_inr} INR')
```

```python
df.to_csv('diamonds_full.csv', encoding='utf-8', index=False)
df.head()
```
